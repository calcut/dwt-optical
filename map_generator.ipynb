Helper code to generate mappings for sensors

Output can be copied into layout or structure_map or surface_map json files to define a new sensor


```python
import string
import json

# A list of 16 abbreviations for surface chemistries, 1 per column
chems = ['MUA',
'DDT',
'NBA',
'MPA',
'PET',
'MBA',
'DTP',
'PFDT',
'MPBA',
'FMT',
'MHOH',
'OT',
'AMP',
'NTP',
'GLU',
'blank']

chem_offset = -3 # Used where the surface chemistries are offset by e.g. -3 as you move down each row. 
x_pos_offset = 1.5 #mm
y_pos_offset = -1.5 #mm
rows = 5
cols = 16
row_list = list(string.ascii_uppercase[:rows]) #to call the rows A,B,C,D etc.
col_list = range(1, cols+1) #to call the columns 01, 02, 03 etc.
surface_map_dict = {}
layout_map_dict = {}

for row in row_list:
    for col in col_list:
        surface_map_dict[F"{row}{col:02d}"] = [chems[col-1], ""] 
        layout_map_dict[F"{row}{col:02d}"] = [x_pos_offset*col, y_pos_offset*row_list.index(row)]
    chems = chems[chem_offset:] + chems[:chem_offset]

print("surface map:")
print(json.dumps(surface_map_dict))
print("layout map:")
print(json.dumps(layout_map_dict))
```
